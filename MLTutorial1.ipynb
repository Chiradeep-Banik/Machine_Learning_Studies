```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
val =1
length = 1000
y = np.zeros(length) 
for i in range (length):
    y[i] = val + random.randrange(-400,400)
    val = val+1

x = np.array([i for i in range(len(y))])

def best_fit_slope(x,y):
    x_mean = np.mean(x)
    y_mean = np.mean(y)
    x2_mean = np.mean(x*x)
    xy_mean = np.mean(x*y)

    m=((x_mean*y_mean)-xy_mean)/((x_mean*x_mean)-x2_mean)
    return m
def best_fit_intercept(x,y):
    x_mean = np.mean(x)
    y_mean = np.mean(y)
    m = best_fit_slope(x,y)
    b = y_mean - (x_mean*m)
    return b
def distance_from_line(y,line_y):
    d = (y-line_y)
    return d

m = best_fit_slope(x,y)
b = best_fit_intercept(x,y)

line = np.zeros(len(x))

for i in range (len(x)):
    line[i] = m*x[i] + b

distances_regrsn =np.zeros(len(x))
distances_mean = np.zeros(len(x))
y_mean = np.mean(y)
for i in range(len(line)):
    distances_regrsn[i]=distance_from_line(y[i],line[i])
    distances_mean[i] = distance_from_line(y[i],y_mean)

se_regsn_line = np.mean(distances_regrsn*distances_regrsn)
se_mean_line = np.mean(distances_mean*distances_mean)

plt.plot(x,line,color="red")
plt.scatter(x,y)
plt.plot(x,[y_mean for a in range(len(x))])
plt.show()
r_squared = 1 - (se_regsn_line/se_mean_line)

print(m,b,r_squared)
```
